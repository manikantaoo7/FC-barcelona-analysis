```python
# Let's say the last column is the target 'Position' that we want to predict
X = foot.iloc[2:, 1:-1]  # Features (all columns except the last one)
y = foot.iloc[2:, -1]   # Target (the last column)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Standardize the feature variables
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the classifier on the training data
clf.fit(X_train_scaled, y_train)

# Predict the positions on the test data
y_pred = clf.predict(X_test_scaled)

# Evaluate the classifier performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Optionally, export the tree visualization
from sklearn.tree import export_graphviz
export_graphviz(clf, out_file='tree.dot', 
                feature_names=X.columns,  
                class_names=label_encoder.classes_,
                filled=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler



# Handling missing values
foot = foot.fillna(foot.mean())

# Drop the 'Player' column as it's categorical and not useful for our analysis
foot.drop(columns=['Player'], inplace=True)

# Excluding the goalkeeper (GK) data from the dataset
foot_no_gk = foot[foot['Positions'] != 'GK']

# Separating the features and the target variable
X_no_gk = foot_no_gk.drop('Positions', axis=1)
y_no_gk = foot_no_gk['Positions']

# Standardizing the features
scaler = StandardScaler()
X_no_gk_scaled = scaler.fit_transform(X_no_gk)

# Splitting the dataset into training and testing sets
X_train_no_gk, X_test_no_gk, y_train_no_gk, y_test_no_gk = train_test_split(X_no_gk_scaled, y_no_gk, test_size=0.2, random_state=0)

# Initializing and training the Naive Bayes classifier
nb_classifier_no_gk = GaussianNB()
nb_classifier_no_gk.fit(X_train_no_gk, y_train_no_gk)

# Making predictions and evaluating the model
y_pred_no_gk = nb_classifier_no_gk.predict(X_test_no_gk)
accuracy_no_gk = accuracy_score(y_test_no_gk, y_pred_no_gk)
report_no_gk = classification_report(y_test_no_gk, y_pred_no_gk)

# Output the accuracy and classification report
print(f"Accuracy: {accuracy_no_gk}\n")
print("Classification Report:\n", report_no_gk)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler


# Preprocessing steps (as you've done before)
foot = foot.fillna(foot.mean())
foot.drop(columns=['Player'], inplace=True)
foot_no_gk = foot[foot['Positions'] != 'GK']
X_no_gk = foot_no_gk.drop('Positions', axis=1)
y_no_gk = foot_no_gk['Positions']

# Standardizing the features
scaler = StandardScaler()
X_no_gk_scaled = scaler.fit_transform(X_no_gk)

# Splitting the dataset
X_train_no_gk, X_test_no_gk, y_train_no_gk, y_test_no_gk = train_test_split(
    X_no_gk_scaled, y_no_gk, test_size=0.2, random_state=0)

# Creating and training the Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=0)
rf_classifier.fit(X_train_no_gk, y_train_no_gk)

# Making predictions
y_pred_rf = rf_classifier.predict(X_test_no_gk)

# Evaluating the model
accuracy_rf = accuracy_score(y_test_no_gk, y_pred_rf)
report_rf = classification_report(y_test_no_gk, y_pred_rf)

print(f"Accuracy: {accuracy_rf}\n")
print("Classification Report:\n", report_rf)
```
